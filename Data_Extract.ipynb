```python
names_content = os.listdir('AFLW2000/')
```

```python
images = [x for x in names_content if x.endswith('.jpg')]
mat = [x for x in names_content if x.endswith('.mat')]
```

```python
def Extract_Data(images, mat, data_frame):   
        x=0
        faceModule = mediapipe.solutions.face_mesh
        for img, mat in zip(images, mat):
            points_data = []
            with faceModule.FaceMesh(static_image_mode=True) as face:
                image = cv2.imread('AFLW2000/' + img)
                results = face.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
                if results.multi_face_landmarks != None:
                    for face in results.multi_face_landmarks:
                        for landmark in face.landmark:
                            x = landmark.x
                            y = landmark.y
                            points_data.append(x)
                            points_data.append(y)
                        points_data = points_data + list(sio.loadmat('AFLW2000/' + mat)['Pose_Para'][0][:3])

                        data.loc[x] = points_data
                        x = x + 1
        return data
```

```python
data = pd.DataFrame(columns=list(range(1, (468*2)+1)) + ['Yaw', 'Pitch', 'Roll'])

final_data = Extract_Data(images, mat, data)
```
